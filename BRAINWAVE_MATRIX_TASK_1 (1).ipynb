```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")

# Load data
# Update this path to the location of your uploaded CSV file in Colab
data_path = '/content/sales_data.csv.csv' # Make sure this path is correct
df = pd.read_csv(data_path)

# Display info and first rows
print("Dataset Shape:", df.shape)
print(df.info())
print(df.head())

# Data Cleaning (remove duplicates as it was in the original code, dropna is not needed for current columns)
df = df.drop_duplicates()


# Analyze customer and location data
# Calculate the number of unique customers
unique_customers = df['customer_unique_id'].nunique()
print(f"\nTotal Number of Unique Customers: {unique_customers}")

# Calculate the distribution of customers by city
city_distribution = df['customer_city'].value_counts()
print("\nTop 10 Cities by Customer Count:")
print(city_distribution.head(10))

# Calculate the distribution of customers by state
state_distribution = df['customer_state'].value_counts()
print("\nDistribution of Customers by State:")
print(state_distribution)

# Visualize location data

# Create a bar plot for the top 10 cities by customer count
plt.figure(figsize=(12, 7))
sns.barplot(x=city_distribution.head(10).index, y=city_distribution.head(10).values, palette="viridis", hue=city_distribution.head(10).index, legend=False)
plt.title('Top 10 Cities by Customer Count')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Create a bar plot for the distribution of customers by state
plt.figure(figsize=(14, 7))
sns.barplot(x=state_distribution.index, y=state_distribution.values, palette="viridis", hue=state_distribution.index, legend=False)
plt.title('Distribution of Customers by State')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAnalysis complete.")
```
